```python
import time
import statistics
import requests
import socket


def portStateChecker():
    print("Port State Checker")
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    hostIpAddress = socket.gethostname()
    ip = socket.gethostbyname(hostIpAddress)
    port_SSH = 22
    port_telnet = 23
    port_DNS = 53
    port_http = 80
    port_https = 443
    ports = [port_SSH,port_telnet,port_DNS,port_http,port_https]
    for port in ports:
        location = (hostIpAddress, port)
        result = sock.connect_ex(location)
        if result == 0:
            print("Port",port, "is open")
        else:
            print("Port",port, "is not open")
        sock.close()
    print(".........")

def connectionStatusChecker():
    print("Connection Status Checker")
    websites = ["http://www.google.com", "http://www.amazon.com"]
    listofTime = []
    for website in websites:
        timeout = 5 # time it takes before deciding you have no internet connection
        start = time.time()
        try:
            request = requests.get(website, timeout=timeout)
            endTime = time.time()-start
            listofTime.append(endTime)
            print("Connected to", website,", it took", endTime,"seconds")
        except (requests.ConnectionError, requests.Timeout) as exception:
            print("No internet connection.")
    if len(listofTime) > 0:
        avgTime = statistics.mean(listofTime)
    else:
        avgTime = 0
    print("Both websites took an average time of : "+str(avgTime)+" seconds")
    assumedMaxTimeForGoodInternetConnection = 2 # in seconds
    if avgTime < assumedMaxTimeForGoodInternetConnection:
        print("You have a good internet connection")
    elif avgTime > assumedMaxTimeForGoodInternetConnection:
        print("You have a poor internet connection") 
    print("............")

justStarted = True
choice = 0
while( justStarted or (choice == 1 or choice == 2 ) ):
    justStarted = False
    print("")
    print("""Network Checker System.
    Please enter your choice from the list of options below:
    1. Port State Checker
    2. Connection Status System
    3. Exit""")
    try:
        
        choice = int(input("Input the check you want to make : "))
        print("")
        if choice == 1:
            portStateChecker()
        elif choice == 2:
            connectionStatusChecker()
        elif choice == 3:
            exit()
        else:
            print("Please input one the options given above")
            choice = int(input("Input the check you want to make : "))
    except:
        print("")
        print("Please input an integer")
        choice = int(input("Input the check you want to make : "))
```
